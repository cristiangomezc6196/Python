```python
from tqdm import tqdm
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures
```

<h2>
    <b>DataSet</b>
</h2>

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod2.csv"
```

```python
df = pd.read_csv(url, header=0)
```

```python
df.head()
```

```python
df.drop(["Unnamed: 0", "Unnamed: 0.1"], axis=1, errors = "ignore", inplace=True)
```

<center>
    <h2>
        <b>Task 1: Using the Cross validation to improve the model</b>
    </h2>
</center>

<i>Divide the dataset into x_data and y_ddata parameters. Here y_data is the "Price" attribute, and x_data has all other attributes in the data set

```python
y_data = df["Price"]
x_data = df.drop("Price", axis="columns")
```

<i>Split the data set into training and testing subsets such that you reserve 10% od the data set for testing purpose</i>

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.10, random_state = 1)
print("Number of test samples :", x_test.shape[0])
print("Number of training samples :", x_train.shape[0])
```

<i>Create a single variable linear regression model using "CPU_frecuency" parameter. Print the R^2 value of this model for the training and testing subsets</i>

```python
lre = LinearRegression()
lre.fit(x_train[["CPU_frequency"]], y_train) 
print(lre.score(x_test[['CPU_frequency']], y_test))
print(lre.score(x_train[['CPU_frequency']], y_train))
```

```python
Rcross = cross_val_score(lre, x_train[['CPU_frequency']], y_train, cv=4)
print("The mean of the folds are", Rcross.mean(), "and the standard deviation is" , Rcross.std())
```

<center>
    <h2>
        <b>Task 2: Overfitting</b>
    </h2>
</center>

<i>Split the data into training and testing componentes again, this time reserving 50% of the data set for testing</i>

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.5, random_state=0)
```

```python
lre = LinearRegression()
Rsqu_test = []
order = [1, 2, 3, 4, 5]
for n in order:
    pr = PolynomialFeatures(degree=n)
    x_train_pr = pr.fit_transform(x_train[['CPU_frequency']])
    x_test_pr = pr.fit_transform(x_test[['CPU_frequency']])    
    lre.fit(x_train_pr, y_train)
    Rsqu_test.append(lre.score(x_test_pr, y_test))
```

```python
plt.plot(order, Rsqu_test)
plt.xlabel('order')
plt.ylabel('R^2')
plt.title('R^2 Using Test Data')
```

<center>
    <h2>
        <b>Task 3: Ridge Regression</b>
    </h2>
</center>

<i>Now consider that you have multiple features. create a polynomial feature model that uses all these parameters with degree=2. Also, create the training and testng attribute sets</i>

```python
pr=PolynomialFeatures(degree=2)
x_train_pr=pr.fit_transform(x_train[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']])
x_test_pr=pr.fit_transform(x_test[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']])
```

<i>Create a Ridge Regression model and evaluate is using values of the hyperparameter alpha ranging from 0.001 to 1 with increments of 0.001. Create a list of all Ridge Regression R^2 scores for training and testing data</i>

```python
Rsqu_test = []
Rsqu_train = []
Alpha = np.arange(0.001,1,0.001)
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha) 
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})
    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

```python
plt.figure(figsize=(10, 6))  
plt.plot(Alpha, Rsqu_test, label='validation data')
plt.plot(Alpha, Rsqu_train, 'r', label='training Data')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.ylim(0, 1)
plt.legend()
```

```python
RR = Ridge()
Grid1 = GridSearchCV(RR, parameters1, cv=4)
```

<i>Find  the Grid Search to the training data</i>

```python
Grid1.fit(x_train[["CPU_frequency", "RAM_GB", "Storage_GB_SSD", "CPU_core", "OS", "GPU", "Category"]], y_train)
```

<i>Print the R^2 score for the test data using the estimator that uses the derived optimum value of alpha.</i>

```python
BestRR = Grid1.best_estimator_
print(BestRR.score(x_test[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core','OS','GPU','Category']], y_test))
```
