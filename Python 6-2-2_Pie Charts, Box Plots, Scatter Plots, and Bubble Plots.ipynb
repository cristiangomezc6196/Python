````python
df_continents = df_can.groupby('Continent').sum()
df_continents.head()
````

<p>Step 2: Plot the data. We will pass in <i>kind = pie</i> keyword, along with the folloring additional parameters</p>

<ul>
    <li><i>autopct:</i> is a string or function used to labe the wedges with their numeric value. The label will be placed inside the wedge. If it is a forat string, the label will be fmt%pct</li>
    <li><i>startangle:</i> rotates the start of the pie chart by agle degrees counterclockside fron the x-axis</li>
    <li><i>shadow:</i> Draws a shadow beneath the pie (to give a 3D feel)</li>
</ul>

````python
df_continents["Total"].plot(kind = "pie",
                            figsize=(5,6),
                            autopct = "%1.1f%%", ## add in porcentages
                            startangle = 90, 
                            shadow = True
                           )

plt.title("Inmigration to Canada by Continent [1980 - 2013]")
plt.axis("equal")
plt.legend(labels=df_continents.index, loc="upper left")

plt.show()
````

<p>Let's make a few modifications to iprove the visuals</p>

<ul>
    <li>Remove the text labels on the pie chart</li>
    <li>Push out the porcentages to sit just outside the pie chart by passing in <i>pctdistance</i> parameter</li>
    <li>Pass in a custom set of colors for continents by passing in <i>colors</i> parameters</li>
    <li><b>explode</b> the pie chart to emphasize the lowest three continnets in <i>explode</i> parameter</li>
</ul>

````python
color_list = ["gold", "yellowgreen", "lightcoral", "lightskyblue", "lightgreen", "pink"]
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] 

df_continents["Total"].plot(kind = "pie", 
                            figsize = (10,6), 
                            autopct = "%1.1f%%",
                            startangle = 90, 
                            shadow = True,
                            labels = None,  #Turn off labels on pie chart
                            pctdistance = 1.12,  #The ratio between the center of each pice slice and the start opf the text generated by autopct
                            colors = color_list,
                            explode = explode_list
                           )

#Scale the title up by 12% to match pctdistance
plt.title("Inmigration to Canada by Continent [1980-2013]", y=1.12, fontsize = 15)

plt.axis("equal")

plt.legend(labels=df_continents.index, loc="upper left", fontsize=7)

plt.show()
````

<p> A <b>box plot</b> is a way of statistically representing the <i>distribution</i> of the data through five main dimensions</i></p>

<ul>
    <li><b>Minimum:</b> The smallest number in the dataset excluding the outliers</li>
    <li><b>First quartile:</b> Middle number between the minimum and the median</li>
    <li><b>Second quartile (Median):</b> Middle number of the sorted dataset</li>
    <li><b>Third quartile:</b> Middle number between <i>median</i>  and <i>maximum</i> </li>
    <li><b>Maximum:</b> The largest number in the dataset excluding the outliers</li>
</ul>

<center>
    <h2>Subplots</h2>
</center>

<p>To visualize multiple p´lots together, we can create a <b>figure</b> and divide it into <b>subplots</b>, each containing a plot. With <b>subplots</b>, we usually work with the <b>artist layer</b> instead of the <b>scripting layer</b></p>

<p>Typical sintax is: </p>

```python
    fig = plt.figure() #Create a figure
    ax = fig.add_subplot(nrows, ncols, plot_number) #Create subplots
```

<p>
    Where
    <ul>
        <li><i>nrows</i> and <i>ncols</i> are used to notionally split the figure into (<i>nrows</i> * <i>ncols</i>) sub-axes</li>
        <li><i>plot_number</i> is used to identify the particular subplot that this function is to creat within the notional grid. <i>plot number</i> starts at 1, increments across rows first and has a maximum of <i>nrows</i> * <i>ncol</i> as show below</li>
</ul>

<p>We can specify wich subplot to place each plot by passing in the <i> ax</i> parameter in <i>plot()</i> method as follows:</p>

````python
fig = plt.figure() # Create figure

ax0 = fig.add_subplot(1, 2, 1) #Add subplot 1 (1 row, 2 columns first plot)
ax1 = fig.add_subplot(1, 2, 2) #Add subplot 2 (1 row, 2 columns, second plot)

#Subplot 1
df_CI.plot(kind ="box",
           color ="blue", 
           vert = False, 
           figsize = (20, 6), 
           ax = ax0 #Add to subplot 1
          )
ax0.set_title("Box Plots of Inmigrants from China  and India (1980 - 2013)")
ax0.set_xlabel("Number of inmigrants")
ax0.set_ylabel("Countries")

#Subplot 2
df_CI.plot(kind = "line",
           figsize = (20,6),
           ax=ax1 #Add to subplot 2
          )
ax1.set_title("Line Plots of inmigrants from China and India (1980 - 2013)")
ax1.set_ylabel("Number of inmigrants")
ax1.set_xlabel("Years")

plt.show()
````

<p>Step 1: Get the dataset. Since we are expecting to use the relationship between years and total population, we will convert years to int type</p>

````python
#We can get the sum() method to get the total population per year
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

#Change the year to type int
df_tot.index = map(int, df_tot.index) 

#Reset the index to put in back as a column in the df_tot dataframe
df_tot.reset_index(inplace=True)

df_tot.columns = ["year", "total"]

df_tot.head()
````

<p>Step 2: Plot the data in Matplotlib by passing in kind = "scatter" as plot argument</p>

````python
df_tot.plot(kind = "scatter",
            x = "year",
            y = "total",
            figsize = (10, 6),
            color = "darkblue"
           )

plt.title("Total Inmigration to Canada fro 1980 - 2013")
plt.xlabel("Year")
plt.ylabel("Number of Inmigrants")

plt.show()
````

<p>Let's try to plot a linear line of best fit</p>

<p>Step 1: Get te equation of line of best fit, We will use Numpy's <i>Polyfit()</i> method </p>

````python
x = df_tot["year"]
y = df_tot["total"]

fit = np.polyfit(x, y, deg=1)

fit
````

<p>Step 2: Plt the rgression line on the <i>scatter plot</i></p>

````python
df_tot.plot(kind= "scatter",
            x="year",
            y="total",
            figsize = (10,6),
            color = "darkblue"
           )
plt.title("Total Inmigration to Canada from 1980 - 2013")
plt.xlabel("Year")
plt.ylabel("Number of Inmigrants")

plt.plot(x, fit[0] * x + fit[1], color = "red")
plt.annotate("y={0:.0f} x + {1:.0f}".format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
````

<p>Let's analyze the effect of this crisis, and compare Argentina's inmigration to that of it's neighbur Brazil. Let's do that using <i>bubble plot</i> of inmigration from Brazil and Argentina for the years 1980 - 2013. We will set the weights for the bubble as the <i>normalized</i> value d the population of each year</p>

<p><b>Step 1:</b> Get the data for Brazil and Argentina. Like i the previous example, we will convert the <i>Years</i> to type int and include it in the dataframe</p>

````python
df_can_t = df_can[years].transpose()

df_can_t.index = map(int, df_can_t.index)

df_can_t.index.name = "Year"

df_can_t.reset_index(inplace=True)

df_can_t.head()
````

<p><b>Step 2:</b> Create a normalized weights</p>

<p>here are several methods of nromalizations in statistics, each with its own use. In this case, se will use <i>featur scaling</i> to bring all values into de range [0, 1]. The general formula is:</p>

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/Module%203/images/Mod3Fig3FeatureScaling.png" align="center">

````python
#normalize Brazil data
norm_brazil = (df_can_t["Brazil"] - df_can_t["Brazil"].min()) / (df_can_t["Brazil"].max() - df_can_t["Brazil"].min())

#normalize Argentina data
norm_argentina = (df_can_t["Argentina"] - df_can_t["Argentina"].min()) / (df_can_t["Argentina"].max() - df_can_t["Argentina"].min())
````

<p><b>Step 3:</b> Plot the data</p>

<ul>
    <li>To plot two different scatter plots in one plot, we can include the axes one plot into the another by passing it via the <i>ax</i> parameter</li>
    <li>We will also pass in the weights using the <i>s</i> parameter. Given that the normalized weights are between 0-1, they won't be visible on the plot. Therefore, we sill:</li>
    <ul>
        <li>multiply weights by 2000 to scale it up on the grapch, and,</li>
        <li>add 10 to compensat for the min value (wich has a 0 eight and therefore scale with x2000</li>
    <ul>
</ul>

````python
#Brazil
ax0 = df_can_t.plot(kind = "scatter",
                    x = "Year",
                    y= "Brazil",
                    figsize = (14, 8),
                    alpha = 0.5,
                    color = "green",
                    s = norm_brazil * 2000 + 10,
                    xlim = (1975, 2015)
                   )

#Argentina
ax1 = df_can_t.plot(kind = "scatter",
                    x = "Year",
                    y= "Argentina",
                    alpha = 0.5,
                    color = "blue",
                    s = norm_argentina * 2000 +10,
                    ax = ax0
                   )

ax0.set_title("Inmigration from Brazil and Argentina from 1980 - 2013")
ax0.set_ylabel("_Number of inmigrants")

ax0.legend(["Brazil", "Argentina"], loc="upper left", fontsize = "x-large")
````
