<h2>Task 3 - GroupBy and Pivot Tables</h2>
<h3>Group the parameters "GPU", "CPU_core", and "Price" to make pivot tables and visualize this connection using the pcolor plot</h3>

<h3>Create the group</h3>

```python
df_gtest = df[["GPU", "CPU_core", "Price"]]
grouped_test1 = df_gtest.groupby(["GPU", "CPU_core"], as_index=False).mean()
print(grouped_test1)
```

<h3>Create the pivot table</h3>

```python
grouped_pivot = grouped_test1.pivot(index="GPU", columns="CPU_core")
print(grouped_pivot)
```

<h3>Create the plot</h3>

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap = "RdBu")

#Label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center 
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

fig.colorbar(im)
```

<h2>Task 4 - Pearson Correlation and p-values</h2>
<h3>Use the scipy.stats.pearsonr() function to evaluate the Pearson Coefficient and the p-value for each parameter tested above.</h3>

```python
for param in ["RAM_GB", "CPU_frequency", "Storage_GB_SSD", "Screen_Size_inch", "Weight_pounds", "CPU_core", "OS",
              "GPU", "Category"]: 
    pearson_coef, p_value = stats.pearsonr(df[param], df["Price"])
    print(param)
    print("The Pearson Correlation Coefficient for ", param, " is: ", pearson_coef, " with a value of P =", p_value)
```
