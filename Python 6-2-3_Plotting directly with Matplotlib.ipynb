<center>
    <h2>Line Plot</h2>
</center>

<p>Let's created a line plot to visualize the inmigrants (to Canada) trend during 1980 to 2013.</p>

<p>We ill create a new datadrame for only columns containing the years, then we will apply sum() on the dataframe</p>

```python
df_line = df_can[years]

total_inmigrants = df_line.sum(axis=0)
total_inmigrants
```

```python
fig, ax = plt.subplots()

ax.plot(total_inmigrants)

ax.set_title("Inmigrants between 1980 - 2013")
ax.set_xlabel("Years")
ax.set_ylabel("Total inmigrants")

plt.show()
```

<p><b>The x label dont have <i>str</i> format and dont looks right. We have to make it like <i>int</i> type</p>

```python
fig, ax = plt.subplots()

total_inmigrants.index = total_inmigrants.index.map(int)

ax.plot(total_inmigrants)

ax.set_title("Inmigrants between 1980 - 2013")

ax.set_xlabel("Years")

ax.set_ylabel("Set inmigrants")

plt.show()
```

<p><b>Let's now customize the above plot's appearance</b></p>

```python
fig, ax = plt.subplots()

total_inmigrants.index = total_inmigrants.index.map(int)

ax.plot(total_inmigrants,
        marker = "s", #squares shapes
        markersize = 5,
        color = "green",
        linestyle = "dotted" #Line style to a dotted line
       ) 

ax.set_title("Inmigrants between 1980 - 2013")

ax.set_xlabel("Years")
ax.set_ylabel("Total Inmigrants")
ax.legend(["Inmigrants"])

plt.show()
```

<p><b>Let's include the background grid, a legend and try to change the limits on the axis</p>

```python
fig, ax = plt.subplots()

ax.plot(total_inmigrants,
        marker = "s",
        markersize = 5,
        color = "green",
        linestyle = "dotted"
       )

ax.set_title("Inmigrants between 1980 to 2013")

ax.set_xlabel("Years")
ax.set_ylabel("Total Inmigrants")

plt.xlim(1975, 2015) 

plt.grid(True)

plt.legend(["Inmigrants"])

plt.show()
```

<center>
    <h2>Histogram</h2>
</center>

<p>Let's find out the frequency distriution of the number (population) of new nmigrants from the various countries to Canada in 2013</p>

```python
try:
    df_country = df_can.groupby(["Country"])["2013"].sum().reset_index()
except:
    print("index reseteado")

fig, ax = plt.subplots(figsize=(10,4))

ax.hist(df_country["2013"])
ax.set_title("New Inmigrants in 2013")
ax.set_xlabel("Number of Inmigrants")
ax.set_ylabel("Number of Countries") 
ax.legend(["Inmigrants"])

plt.show()
```

<p>Our plot doesn't match with the bars</p>

<p>The <i>hist</i> function returns list of arrays with 1. counts and 2. bins. We can fetch that using unpacking functionality and further use the bins as x-ticks</p>

```python
fig, ax = plt.subplots(figsize =(10,4))
count = ax.hist(df_country["2013"])

ax.set_title("New Inmigrants in 2013")
ax.set_xlabel("Number of inmigrants")
ax.set_ylabel("Number of Countries")
ax.set_xticks(list(map(int, count[1])))
ax.legend(["Inmigrants"])

plt.show()
```

<p><b>We can also plot umtiple histograms on the same plot</b></p>

<p><b>What's the inmigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013</p>

```python
df = df_can.groupby(["Country"])[years].sum()
df_dns = df.loc[["Denmark", "Norway", "Sweden"], years]
df_dns = df_dns.transpose()
df_dns.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.hist(df_dns)
ax.set_title("Inmigration from Denmark, Norway, and Sweden from 1980 - 2013")
ax.set_xlabel("Number of inmigrants")
ax.set_ylabel("Number of years")

ax.legend(["Denmark", "Norway", "Sweden"])

plt.show()
```

<p><b>Question:</b> What is the inmigration distribution for china and India for years 2000 to 2013</p>

```python
df = df_can.groupby(["Country"])[years].sum()

y = list(map(str, range(2000,2014)))

df_ci = df.loc[["China", "India"], y]
df_ci = df_ci.transpose()

fig, ax = plt.subplots()
ax.hist(df_ci)
ax.set_title("Inmigration fron denmark, Norway, adn Sweden from 1980 - 2013")
ax.set_xlabel("Number of inmigrants")
ax.set_ylabel("Number of Years")
ax.legend(["China", "India"])

plt.show()
```

<p>Let's create a pie chart representing the "Total inmigrants" for the year 1980 - 1985</p>

<p><b>Question:</b> Create a pie chart representing the total inmigrants proportion for each continent</p>

<p>First we will have to group the data over cntinent and get the sum on total. Then you can pass this data to the pie function</p>

```python
df_can.head()
df_con = df_can.groupby("Continent")["Total"].sum().reset_index()
print("index reseteado")

label = list(df_con.Continent)

label[3] = "LAC"
label[4] = "NA"

df_con
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.pie(df_con["Total"],
       colors = ["gold", "blue", "lightgreen", "coral", "cyan", "red"],
       autopct = "%1.1f%%",
       pctdistance = 1.25
      ) 

ax.set_aspect("equal")
ax.set_title("Continent-wise distribution of inmigrants")
ax.legend(label, bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

<center>
    <h2>Sub-Plotting</h2>
</center>

<p>Let's create a line and scatter plot in one row</p>

```python
fig, axs = plt.subplots(1,2, sharey=True)

axs[0].plot(total_inmigrants)
axs[0].set_title("Line Plot in inmigrants")

axs[1].scatter(total_inmigrants.index, total_inmigrants)
axs[1].set_title("Scatter plot on inmigrants")

axs[0].set_ylabel("Number of inmigrats")

fig.suptitle("Subploting Example", fontsize=15)

fig.tight_layout()

plt.show()
```

<p>You can also implement the subpotting with <b>add_subplot()</b> as below:</p>

```python
fig = plt.figure(figsize=(8,4))

axs1 = fig.add_subplot(1,2,1)

axs1.plot(total_inmigrants)
axs1.set_title("Line plot on inmigrants")

axs2 = fig.add_subplot(1,2,2)
axs2.barh(total_inmigrants.index, total_inmigrants)
axs2.set_title("Bar plot on inmigrants")

fig.suptitle("Subplotting Example", fontsize=15)

fig.tight_layout()

plt.show()
```
