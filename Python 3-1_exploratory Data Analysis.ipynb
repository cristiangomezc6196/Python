<h3>Basics of Grouping</h2>

```python
df["drive-wheels"].unique()
```

```python
df_group_one = df[["drive-wheels", "body-style", "price"]]
```

<h3>Add 3 columns: mean, max and min</h3>

```python
df_grouped = df_group_one.groupby(["drive-wheels"], as_index = False).agg({"price":["mean", "max", "min"]})
df_grouped
```

<h3>Just add one column: mean</h3>

```python
df_gptest = df[["drive-wheels", "body-style", "price"]]
grouped_test1 = df_gptest.groupby(["drive-wheels", "body-style"], as_index = False).mean()
grouped_test1
```

<h3>Pivot</h3>

```python
grouped_pivot = grouped_test1.pivot(index = "drive-wheels", columns = "body-style")
grouped_pivot
```

<h3>When we don't have data for some of the pivot cells. We can fill these missing cells with the value 0, but any other value could potentially be used as well</h3>

```python
grouped_pivot = grouped_pivot.fillna(0)
```

```python
grouped_pivot
```

<h3>Variables: Drive Wheels and Body Style vs Price</h3>

```python
plt.pcolor(grouped_pivot, cmap="RdBu")
plt.colorbar() 
plt.show()
```

<h3>The default labels convey no useful information to us. Let's change that</h3>

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

<h2>Wheel-base vs Price</h2>

```python
pearson_coef, p_value = stats.pearsonr(df["wheel-base"], df["price"])
print("The Pearson Correlation Coefficient is: ", pearson_coef, " with a P-Value of P =", p_value)
```

<h3>The p-value< 0.001, the correlation between horsepower and price is statiscally significant, and the linear relationship is quite strong (close to 1)</h3>
