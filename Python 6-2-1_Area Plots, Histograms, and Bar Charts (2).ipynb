<p>Before we proceed with creating the histogram plot, let's first examine the data split into intervals. To do this, we will us <b>Numpy's</b> <i>histogram</i> method to get the bin ranges and frecuency cous as follows</p>

```python
#Let's quickly view the 2013 data
df_can[["2013"]].head()
```

```python
#np.histogram returns 2 values

count, bin_edges = np.histogram(df_can["2013"])

print(count)
print(bin_edges)
```

<p>By default the <i>histogram</i> method braks uo the dataset into 10 bins</p>

<p>We can easly graph this distribution by passing kind=hist to plot()</p>

```python
df_can['2013'].plot(kind='hist', figsize=(8, 5))

plt.title("Histogram of inmigration from 195 Countries in 2013")
plt.ylabel("Number of Countries")
plt.xlabel("Number of imigrants")

plt.show()
```

<p>The x-axis labels do not match with the bin size. This can be fixed by passing in a xticks keyword that contains the list of the bin sizes</p>

```python
#"bin_edges" is a list of bin intervals

count, bin_edges = np.histogram(df_can["2013"])

df_can['2013'].plot(kind='hist',
                    figsize=(8, 5), 
                    xticks=bin_edges)

plt.title("Histogram of inmigration from 195 countries in 2013")
plt.ylabel("Number of Countries")
plt.xlabel("Number of Inmigrants")

plt.show()
```

```python
df_can.loc[["Denmark", "Norway", "Sweden"], years] 
```

```python
#Generate the histogram 
df_can.loc[["Denmark", "Norway", "Sweden"],years].plot(kind = "hist")
```

<b>Don't look right...</b>

```python
#Tranpose dataframe
df_t = df_can.loc[["Denmark", "Norway", "Sweden"], years].transpose()
df_t.head()
```

```python
#Generate histogram

df_t.plot(kind = "hist",
          figsize = (10, 6))

plt.title("Histogram of Inmigration fron Denmark, Norway, and Sweden from 1980 - 2013")
plt.ylabel("Number of years")
plt.xlabel("Number of inmigrants")
```

<p>Let's make a few modifications to improve the impact and aesthetics of the previous plot:</p>

<ul>
    <li>increase the bin size to 15 by passing in bins parameters</li>
    <li>set transparency to 60% by passing in alpha parameter</li>
    <li>label the x.axis by apssing in x-label parameter</li>
    <li>change the color of the plots b passing in color parameter</li>
</ul>

```python
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind = "hist",
          figsize=(10, 6),
          bins = 15,
          alpha = 0.6,
          xticks = bin_edges, 
          color = ["coral", "darkslateblue", "mediumseagreen"]
         )

plt.title("Histogram of inmigration from Denmark, Norway, and Sweden from 1980 - 2013")
plt.ylabel("Number of years")
plt.xlabel("Number of inmigrants")

plt.show()
```

<p>The available colors:</p>

```python

for name, hex in mpl.colors.cnames.items():
    print(name, hex)
```

<p>If we dont want the plots to overlap each other, we can stack thenm using the stacked parameter. Let's also adjust the min and max x-axis labels to remove the extra gap on the edges of the plot. We ca pass a tuple (min, max) using the xlim paramter, as show below</p>

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] -10 #First bin value
xmax = bin_edges[-1] + 10 #Last bin value 

#Stacked histogram
df_t.plot(kind="hist",
         figsize=(10,6),
         bins=15, 
         xticks = bin_edges,
         color = ["coral", "darkslateblue", "mediumseagreen"],
         stacked = True,
         xlim = (xmin, xmax)
        )

plt.title("Histogram of inmigration from Denmark, Norway, and Sweden rom 1980 - 2013")
plt.ylabel("Number of Years")
plt.xlabel("Number of inmigrants")

plt.show()
```

<p><b>Question:</b> Let's compare the number of Iceland inmigrants (country = "Iceland") to Canada from year 1980 to 2013</p>

```python
df_iceland = df_can.loc[["Iceland"], years].transpose()
df_iceland.head()
```

```python
#Plot the data

df_iceland.plot(kind = "bar", 
                figsize = (10,6)
               )

plt.title("Icelandic inmigrants to Canada fro 1980 to 2013")
plt.xlabel("Year") 
plt.ylabel("Number of inmigrants")

plt.show()
```

<p>The bar plot above shows the total number of inigrants broken down by ach year. We can clearly see the impact if the financial crisis; the number if inmigrants to Canada started increasing rapidly after 2008</p>

<p>Let's annotate this on the plot using the <i>annotate</i> method of the scripting later or the pyplot interface. We sill pass in the following parameters:</p>

<ul>
    <li><b>s</b>: str, the text of annotation </li>
    <li><b>xy</b>: Tuple specifyin the (x,y) point to place the text (in this case, end point of arow)</li>
    <li><b>xytext</b>: Tuple specifying the (x,y) point to place the text (in this case, start point of arrow)</li>
    <li><b>xycords</b>: The cordinate system that xy is given in - 'data' uses the coordinate system of the object being annotated (default)</li>
    <li><b>arrowprops</b>: Takes a dicionary of properties to draw the arrow</li>
        <ul>
            <li><b>arrowstyle</b>: Specifies the arrow style <i>-></i> is standar arrow</li>
            <li><b>connectionstyle</b>: Specifies the connection type. <i>arc3</i> is a straight line</li>
            <li><b>color</b>: Specifies color of arrow</li>
            <li><b>lw</b>: Specifies the line width</li>
        </ul>
</ul>

```python
df_iceland.plot(kind='bar', figsize=(10, 6))  # rotate the xticks(labelled points on x-axis) by 90 degrees

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('',  # s: str. Will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='red', lw=2)
             )

plt.show()
```

<p>Let's annotate a text to go over the arrow. We will pass in the following additional paramters: </p>

<ul>
    <li><b>Rotation</b>: Rotation angle of text in degrees (counter clockwise)</li>
    <li><b>va</b>: Vertical alignment of text ["center", "top", "bottom", "baseline"]</li>
    <li><b>ha</b>: Horizontal alignment of text ["center", "right", "left"]</li>
</ul>

```python
df_iceland.plot(kind='bar', figsize=(10, 6), rot=90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('',  # s: str. will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis',  # text to display
             xy=(28, 30),  # start the text at at point (year 2008 , pop 30)
             rotation=72.5,  # based on trial and error to match the arrow
             va='bottom',  # want the text to be vertically 'bottom' aligned
             ha='left',  # want the text to be horizontally 'left' algned.
             )

plt.show()
```
