<h4><b>Select Column</b></h4>
<p></p><b>There're two ways to filter on a column name</b></p>

<p>Only works if the column name does not have spaces or special characters</p>
<ul>
    <li>df.column_name --> Returns series</li>

<p>More robust, and can filter on multiple columns</p>
<ul>
    <li>df["column"] --> Returns series</li>

</ul>

<h3><b>Filtering based on a criteria</b></h3>
<p>To filter the dataframe based on a condition, we simply pass the condition as a boolean vector</p>
<p>For example, let's filter the dataframe to whos the data on Asian Countries (AreaName = Asia)</p>

```python
condition = df_can["Continent"] == "Asia"
print(condition)
```

```python
df_can[condition]
```

<p>We can pass multiple criteria in the same line</p>

```python
df_can[(df_can["Continent"] == "Asia") & (df_can["Region"] == "Southern Asia")] 
```

<h3><b>Sorting values of a DataFrame or series</b></h3>

<p>You can use the <b>sort_values()</b> function is used to sort a DataFrame or a Seires based on one or more columns</p>
