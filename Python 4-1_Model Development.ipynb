<center>
    <h2>2. Polynomial Regression and Pipelines</h2>
</center>

<i><b>Polynomial regression </b>is a particular case of the general linear regression model or multiple linear regression models</i>
<p><i>There are different order of polynomial regression: </i></p>

<center><b>Quadratic - 2nd Order</b></center>
$$
Yhat = a + b_1 X +b_2 X^2 
$$


<center><b>Cubic - 3rd Order</b></center>
$$
Yhat = a + b_1 X +b_2 X^2 +b_3 X^3\\\\\\
$$


<center><b>Higher-Order</b>:</center>
$$
Y = a + b_1 X +b_2 X^2 +b_3 X^3 ....\\
$$

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

<i>Let's get the variables</i>

```python
x = df["highway-mpg"]
y = df["price"]
```

<i>Let's fit the polynomial using the function polyfit then use the function poly1d to display the polynomial function</i>

```python
f=np.polyfit(x,y,3)
p=np.poly1d(f)
print(p)
```

<i>Let's plot the function</i>

```python
PlotPolly(p,x,y,"highway-mpg")
```

```python
np.polyfit(x,y,3)
```

<i>We can already see from plotting that this polynomial nodel performs better than the linear model. This is because the generated polynomial function "hits" more of the data points</i>

    <h2>4. Measures for In-Sample Evaluation</h2>
</center>

<p>When evaluating our models, not only do we want to visualize the results, but we also want a quantitative measure to determine how accurate the model is.</p>
<p>Two very important measures that are often used in Statistics to determine the accuracy of a model are:</p>
<ul>
    <li><b>R^2/R-squared</b></li>
    <li><b>Mean Squared Error (MSE)</b></li>
</ul>
<p><b>R-squared</b></p>
<p>Also know as the coeffficient of determination. Determine how close that data is to the fitted regression line</p>
<p><b>MEan Squared Error (MSE)</b></p>
<p>Measures the average of the squares of errors. That's the diference between actual value(y) and the estimated value(y)</p>
